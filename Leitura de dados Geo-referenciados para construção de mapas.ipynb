```python
import pandas as pd
import io 
import requests as req

url = 'https://github.com/kelvins/Municipios-Brasileiros/blob/main/csv/municipios.csv'

s = req.get(url).content
s
```

```python
file_name = io.StringIO(s.decode('utf-8'))
file_name
```

```python
df = pd.read_html(file_name)
df = df[0]
df
```

# Tratamento dos dados

```python
# opção de mudar os valores da coluna
dfk = df.copy()
dfk.iloc[:,0] = df.iloc[:,0].fillna(0) 
dfk
```

```python
# opção de excluir a coluna
dfk = dfk.drop(['Unnamed: 0'], axis=1)
dfk
```

```python
# Opção com dropna
df2 = df.copy()
df2 = df2.transpose()
df2 = df2.dropna()
df2 = df2.transpose()
df2
```

```python
print(dfk.columns.ravel())
```

# Selecionando dodos para DataFrame 

```python
dfk = dfk.loc[df['codigo_uf']==35, ['nome', 'latitude', 'longitude']]
dfk
```

# Criando mapa com pacot Folium

```python
import folium as f

map_ = f.Map(location = [-22.7864889, -50.6786708], zoom_start=7)
for i in range(0, len(dfk)):
    pop = str(dfk.iloc[i]['nome'])
    pop = pop.encode('latin1')
    f.Marker([dfk.iloc[i]['latitude'], dfk.iloc[i]['longitude']], popup=pop).add_to(map_)

map_
```

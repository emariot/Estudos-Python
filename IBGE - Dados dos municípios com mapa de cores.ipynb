```python
import requests as req
import pandas as pd 
headers = {
    'Content-Type': 'application/json;charset=UTF-8',
    'User-Agent': 'google-colab',
    'Accept': 'application/json, text/plain, */*',
    'Accept-Encoding': 'gzip, deflate, br',
    'Accept-Language': 'pt-BR, pt;q=0.9,en-US;q=0.8,en;q=0.7',
    'Connection': 'keep-alive'
}
url = 'https://servicodados.ibge.gov.br/api/v1/localidades/estados/35/municipios'
d = req.get(url, headers=headers).json()
d
```

## Criando DataFrame


```python
mun_sp = [municipios['nome'] for municipios in d]
id_sp = [municipios['id'] for municipios in d]

data_sp = pd.DataFrame.from_dict({
    "ID":id_sp,
    "MUNICÍPIOS":mun_sp
})

data_sp
```

## Extraindo indicador de tabela CSV 

```python
ind = pd.read_csv(r'C:\Users\Mariot\Documents\PythonFundamentos\Estudos Python\datasets\ISDEL_Cidades_Estado_SPIBGE.csv', encoding='latin-1', sep=';', decimal=',', low_memory=False).dropna()
ind
```

```python
data = pd.merge(left=data_sp, right=ind, left_on='ID', right_on='Cód_Municipio', how='outer')
data
```

```python
data.columns = data.columns.str.lower().str.replace(" ", "").str.replace("ó","o").str.replace('í', 'i')
print(data.columns.ravel())
```

## Buscando os dados Georeferenciados

```python
url = 'https://servicodados.ibge.gov.br/api/v2/malhas/35/?formato=application/vnd.geo+json&resolucao=5'
geo_json = req.get(url, headers=headers).json()
```

## Criando uma linha para a métrica no documento json

```python
for m in geo_json['features']:
    cod = m['properties']['codarea']
    m['properties']['isdel'] = float(data.loc[int(cod), 'isdel'])

                                                
```

```python
geo_json
```

```python
geo_json['features'][0]['properties']['isdel']
```

```python
import folium as fl
import numpy as np



map_ = fl.Map(
    location=[-22.7864889, -49.6786708], zoom_start=7, 
    tiles='Stamen Terrain')
fgp = fl.FeatureGroup(name='isdel')
fgp.add_child(fl.GeoJson(data=geo_json, 
                         style_function=lambda x:{'fillColor':'light_green' if x['properties']['isdel']<0.30
                             else 'yellow' if 0.30 <= x['properties']['isdel']<0.40
                             else 'orange' if 0.40 <= x['properties']['isdel']<0.50
                             else 'red' if 0.50 <= x['properties']['isdel']<0.60
                             else 'white'
                         }))

map_.add_child(fgp)
map_.add_child(fl.LayerControl())

map_
```

```python
map_ = fl.Map(
    location=[-22.7864889, -49.6786708], zoom_start=7, 
    tiles='stamentoner')
fgp = fl.FeatureGroup(name='isdel')
fgp.add_child(fl.GeoJson(data=geo_json, 
                         style_function=lambda x:{
                             'fillColor':colormap(x['properties']['isdel']),
                             'fillcolor':'black',
                             'fillOpacity':0.5,
                             'weight':0.8
                             
                         }))

map_.add_child(fgp)
map_.add_child(fl.LayerControl())
colormap.caption = ("ISDEL em 2019")
colormap.add_to(map_)
map_
```

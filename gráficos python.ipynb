```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from random import randint
```

```python
#Gr치fico de Barras

car_1 = [15, 40, 28, 24, 12]
car_2 = [40, 28, 40, 26, 35]

index = np.arange(5)
y = np.arange(0, 110, 10)
width = 0.4

plt.xticks(index, ('Day_1', 'Day_2', 'Day_3', 'Day_4', 'Day_5'))

bar_1 = plt.bar(index, car_1, width, color = 'r')
bar_2 = plt.bar(index, car_2, width, bottom = car_1, color='g')

plt.ylabel('Sales')
plt.yticks(y)

plt.title('Sales by car type')

plt.legend((bar_1[0], bar_2[0]), ('Car_1', 'Car_2'))

plt.show()
```

```python
#Gr치fico de Pizza

labels = 'Car1', 'Car2', 'Car3', 'Car4'
size = [10, 25, 50, 15]
explode = (0, 0.1, 0, 0)

plt.pie(size, labels=labels, autopct='%1.1f%%', explode=explode)

plt.show()
```

```python
# Gr치fico de Linha 
x = np.arange(11)
y = np.random.randint(10, 30, size=11)

fig, ax = plt.subplots()

ax.plot(x,y)

ax.set(xlabel='Eixo(x)', ylabel='Eixo(y)', title='Gr치fico de Custo')


ax.grid()

plt.show()
```

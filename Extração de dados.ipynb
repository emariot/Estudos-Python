# Criando uma função para extrair dados

```python
import pandas as pd
import io
import requests

header = {
    "User-Agent": "Mozilla/5.0 (Windows NT 6.1; Win64; x64; rv:47.0) Gecko/20100101 Firefox/47.0",
    "X-Requested-With": "XMLHttpRequest"
}

def readYield(name):
    print("Leitura dos dados da empresa: "+name)
    filename = base_url + str(name)
    r = requests.get(filename, headers=header)
    dt = pd.read_html(r.text)

    yi = 0.0

    if(len(dt)>1):
        tab = dt[2]
        m,n = tab.shape
        if(m > 7):
            if(n > 2):
                yi = tab.iloc[8,3]
            else:
                print('Faltou o dado_3')
        else:
            print('Faltou o dado_2')
    else:
        print('Faltou o dado_1')
    return yi


# Leitura dos códigos das empresas
url = "https://www.fundamentus.com.br/detalhes.php"
r = requests.get(url, headers=header)
d = pd.read_html(r.text)
df = d[0]

base_url="https://www.fundamentus.com.br/detalhes.php?papel="
name = list(df["Papel"])
comercial = list(df["Nome Comercial"])

dk = pd.DataFrame(columns=['Nome', 'Código', 'Yield'])

for i in range(0, len(name)):
    yi = readYield(name[i])
    dk.loc[i] = [comercial[i], name[i],yi]
dk
```

# Limpeza dos dados

```python
dkt = dk.astype(str)
dkt['Yield'] = dkt['Yield'].str.replace('%','')
dkt['Yield'] = dkt['Yield'].str.replace('.','')
dkt['Yield'] = dkt['Yield'].str.replace(',','.')
dkt['Yield'] = dkt['Yield'].astype(float)
dkt.sort_values(by=['Yield'], inplace=True, ascending=False)
dkt.head(30)
```

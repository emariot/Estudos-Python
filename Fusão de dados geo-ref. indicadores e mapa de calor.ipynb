```python
import pandas as pd
import io
import requests as req
```

```python
# Dataset Municípios

url = 'https://github.com/kelvins/Municipios-Brasileiros/blob/main/csv/municipios.csv'
s = req.get(url).content
file = io.StringIO(s.decode('utf-8'))
d = pd.read_html(file)
dfk = d[0]
dfk
```

```python
dfk = dfk.drop('Unnamed: 0', axis=1)
dfk
```

```python
# Tabela apenas com variáveis necessárias

dfk = dfk.loc[dfk['codigo_uf'] == 35,['codigo_ibge', 'nome', 'latitude', 'longitude']]
dfk
```

```python
# Dataset Métricas

dfm = pd.read_csv(r'C:\Users\Mariot\Documents\PythonFundamentos\Estudos Python\datasets\ISDEL_Cidades_Estado_SPIBGE.csv', encoding='latin-1', sep=';', decimal=',', low_memory=False)
dfm
```

```python
dfm = dfm.dropna()
dfm
```

```python
dfm.columns = dfm.columns.str.lower().str.replace(" ","").str.replace("ó","o")
```

```python
print(dfm.columns.ravel())
```

```python
dfm
```

```python
# Fusão entre tabela de municípios e as métricas: dfk e dfm

dff = dfk.merge(dfm, left_on='codigo_ibge', right_on="cod_municipio")
dff
```

```python
print(dff.columns.ravel())
```

# Montagem do Mapa de Calor

```python
import folium as fl
from folium.plugins import HeatMap

map_ = fl.Map(location=[-22.7864889, -50.6786708], zoom_start=7)
total_data = dff[['latitude', 'longitude','isdel']].groupby(['latitude','longitude']).sum().reset_index().values.tolist()
grad = {'0.2':'Navy', '0.30':'Blue', '0.35':'Green', '0.45':'Yellow', '0.6':'Red', '0.8':'Purple'}

HeatMap(total_data, radius=8, gradient=grad, max_zoom=13).add_to(map_)

map_
```

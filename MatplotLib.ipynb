```python
import matplotlib.pyplot as plt
```

# Plot and Scatter

```python
x = [1, 2, 3, 4, 5, 6]
y = [1, 4, 9, 16, 25, 38]

plt.plot(x,y, color='b')
plt.scatter(x, y, color='r')
plt.title('Primeiro Gr√°fico com MatplotLibe')
plt.ylabel('Eixo Y')
plt.xlabel('Eixo X')
plt.grid()
plt.legend(['Previsto','Verificdo'])
plt.show()
```

# MARKER

```python
plt.plot(x,y,'r-.')
plt.scatter(x,y, marker='*', color='b')
plt.show()
```

# Legend

```python
#x = [1, 2, 3, 4, 5, 6]
#y = [1, 4, 9, 16, 25, 38]
y2 = [ 4, 6, 9, 8, 20, 28]
plt.plot(x,y, color='b', label='Produto')
plt.plot(x,y2, color='r', label='Demanda')
plt.legend()
plt.show()
```

# Plot plus Bar

```python
a = list(range(1, 10, 1))
b = [1, 3, 5, 3, 1, 3, 5, 6, 1]
c = [2, 4, 6, 4, 2, 4, 6, 2, 4]
plt.plot(a, b, 'b--')
plt.bar(a, c, color='gray')
```

# Two bars


```python
x2 = [1, 2, 3, 4, 5, 6, 7, 8, 9]
x3 = [6, 7, 8, 9, 10, 11, 12, 13, 14]
y2 = [1, 3, 5, 3, 1, 3, 5, 3, 1]
y3 = [2, 4, 6, 4, 2, 4, 6, 4, 2]
plt.bar(x2, y2, label='Empresa 1', color='b')
plt.bar(x3, y3, label='Empresa 2', color='r')
plt.legend()
plt.show()
```

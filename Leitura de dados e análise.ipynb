# Importação de arquivos .csv na Internet

```python
import io
url = "https://covid.ourworldindata.org/data/ecdc/total_cases.csv"
s = requests.get(url).content
df = pd.read_csv(io.StringIO(s.decode('utf-8')))
df
```

```python
dft = df.copy()
dft = dft.transpose()
dft
```

```python
#dft = dft.dropna()
dft = dft.fillna(0)
dft
```

```python
import matplotlib.pyplot as plt

x = df["date"]
y1 = df["Italy"]
y2 = df["Spain"]

plt.plot(x, y1, label='Itália')
plt.plot(x, y2, label='Espanha')
plt.legend()
plt.title("Evolução de Casos Confirmados")
plt.show()


```

```python
x = [*range(1, len(dft.iloc[1])+1,1)]
y3 = dft.iloc[1]
plt.plot(x,y1,label='Itália')
plt.plot(x,y2,label='Espanha')
plt.plot(x,y3,label='Mundo')
plt.legend()
plt.title('Evolução de Casos Confirmados')
plt.show()
```

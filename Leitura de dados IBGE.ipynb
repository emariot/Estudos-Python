# Fonte de dados - https://servicodados.ibge.gov.br/api/docs

```python
import requests
headers = {
    'Content-Type': 'application/json;charset=UTF-8',
    'User-Agent': 'google-colab',
    'Accept': 'application/json, text/plain, */*',
    'Accept-Encoding': 'gzip, deflate, br',
    'Accept-Language': 'pt-BR, pt;q=0.9,en-US;q=0.8,en;q=0.7',
    'Connection': 'keep-alive'
}

municipios_json = requests.get('https://servicodados.ibge.gov.br/api/v2/malhas/35/?formato=application/vnd.geo+json&resolucao=5', headers=headers)
municipios_json = municipios_json.json()
municipios_json
```

```python
import folium as fl

mapa_sp = fl.Map(location=[-22.7864889, -50.6786708], zoom_start=7)

fl.GeoJson(municipios_json).add_to(mapa_sp)
mapa_sp
```

```python
for municipios in municipios_json['features']:
    cod = municipios['properties']['codarea']
    centroide = municipios['properties']['centroide']
    print("Codearea = "+str(cod)+" ---- Centroide"+str(centroide))
```
